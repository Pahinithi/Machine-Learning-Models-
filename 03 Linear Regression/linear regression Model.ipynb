```python
input_data = car_info.drop(columns = ["car_name" , "selling_price"])
out_data = car_info ["selling_price"]
```

```python
print (input_data.shape)
print (out_data.shape)
```

```python
print (input_data)
print (out_data)
```

# change number simliar range of number (StandardScaler)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
input_data = scaler.fit_transform(input_data)
print (input_data)
```

# Model test (allocation datas test,train)

```python
from sklearn.model_selection import train_test_split 
input_data_train , input_data_test , output_data_train , output_data_test = train_test_split (input_data , out_data , test_size = 0.3)
```

```python
print(input_data_train.shape)
print(output_data_train.shape)
print(input_data_test.shape)
print(output_data_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
model = LinearRegression()
model.fit(input_data_train , output_data_train)
```

# predict the data

```python
predited_sellingprice = model.predict(input_data_test)
```

# commpare predit value and assign value 

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
print("Mean Absolute Error:" , (metrics.mean_absolute_error(predited_sellingprice, output_data_test)))
print("Mean Squared Error:" , (metrics.mean_squared_error(predited_sellingprice, output_data_test)))
print("R2 Score:" , (metrics.r2_score(predited_sellingprice, output_data_test)))
```
